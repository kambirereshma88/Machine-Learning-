```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score 
```

# Data Collection and Processing

```python
loan_dataset=pd.read_csv("Loan_dataset.csv")
```

```python
loan_dataset
```

```python
# statistical measures
loan_dataset.describe()
```

```python
#number of missing values in each column
loan_dataset.isnull().sum()
```

```python
#dropping all missing values 
loan_dataset=loan_dataset.dropna()
```

```python
loan_dataset.isnull().sum()
```

#Label Encoding

```python
#label encoding    #this is dictionary dataset where yes = 1 and n = 0
loan_dataset.replace({"Loan_Status":{"N":0,"Y":1}},inplace=True)
```

```python
#Printing the first 5 rows of dataframe to check Loan_staus value 

loan_dataset.head()
```

```python
#Dependent column values

loan_dataset['Dependents'].value_counts()
```

```python
#replacing the value of 3+ to 4

loan_dataset = loan_dataset.replace('3+',4)
```

```python
#Dependent column values

loan_dataset['Dependents'].value_counts()
```

# Data Visualization

```python
#education and loan status

sns.countplot(data=loan_dataset,x="Education",hue="Loan_Status")
```

```python
#martial status and loan Status
# if a person is married then loan is approved so that any one from both can contribute to loan  
sns.countplot(data=loan_dataset,x="Married",hue="Loan_Status")
```

```python
# gender and loan status
sns.countplot(data=loan_dataset,x="Gender",hue="Loan_Status")
```

```python
#convert categrical columns to numerical values

loan_dataset.replace({"Married":{"No":0,"Yes":1},"Gender":{"Male":1,"Female":0},"Self_Employed":{'No':0 ,'Yes':1},'Property_Area':{
    'Rural':0,'Semiurban':1,'Urban':2},'Education':{'Graduate':1,'Not Graduate':0}},inplace=True)
```

```python
# here we have converted all categorical data into numerical data
loan_dataset.head()     
```

```python
#we will separate all data  and label   
#separating the data and  label 
x=loan_dataset.drop(columns=['Loan_ID','Loan_Status'],axis=1)
y=loan_dataset['Loan_Status']
```

```python
x
```

```python
y
```

# Splitting of data into training and testing 


```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
#splited data into training and testing data
print(x.shape,x_train.shape,x_test.shape)
```

# Training model

# Applying algorithm

# Applying Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import classification_report, accuracy_score

# Train and evaluate Logistic Regression
logistic_model = LogisticRegression(random_state=42)
logistic_model.fit(x_train, y_train)
y_pred_logistic = logistic_model.predict(x_test)

print("Logistic Regression")
print("Accuracy:", accuracy_score(y_test, y_pred_logistic))
print(classification_report(y_test, y_pred_logistic))
```

# Applying Support Vector Machine Model

```python
from sklearn.svm import SVC

# Train and evaluate SVM
svm_model = SVC(random_state=42, kernel='linear')
svm_model.fit(x_train, y_train)
y_pred_svm = svm_model.predict(x_test)

print("Support Vector Machine")
print("Accuracy:", accuracy_score(y_test, y_pred_svm))
print(classification_report(y_test, y_pred_svm))
```

# Applying DecisionTreeClassifier Model

```python
from sklearn.tree import DecisionTreeClassifier

# Train and evaluate Decision Tree
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(x_train, y_train)
y_pred_dt = dt_model.predict(x_test)

print("Decision Tree")
print("Accuracy:", accuracy_score(y_test, y_pred_dt))
print(classification_report(y_test, y_pred_dt))
```

# Applying RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier

# Train and evaluate Random Forest
rf_model = RandomForestClassifier(random_state=42, n_estimators=100)
rf_model.fit(x_train, y_train)
y_pred_rf = rf_model.predict(x_test)

print("Random Forest")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```

# Applying GradientBoostingClassifier

```python
from sklearn.ensemble import GradientBoostingClassifier

# Train and evaluate Gradient Boosting
gb_model = GradientBoostingClassifier(random_state=42)
gb_model.fit(x_train, y_train)
y_pred_gb = gb_model.predict(x_test)

print("Gradient Boosting")
print("Accuracy:", accuracy_score(y_test, y_pred_gb))
print(classification_report(y_test, y_pred_gb))
```

# Accuracy Scores for all Models

```python
# Store accuracy scores for all models
results = {
    "Logistic Regression": accuracy_score(y_test, y_pred_logistic),
    "Decision Tree": accuracy_score(y_test, y_pred_dt),
    "Random Forest": accuracy_score(y_test, y_pred_rf),
    "Support Vector Machine": accuracy_score(y_test, y_pred_svm),
    "Gradient Boosting": accuracy_score(y_test, y_pred_gb),
}

# Print results
print("Model Comparison:")
for model, accuracy in results.items():
    print(f"{model}: {accuracy:.2f}")
```

# INSIGHTS

Best Algorithm: The algorithm with the highest accuracy and F1-score are Gradient Boosting or Random Forest for given data.
